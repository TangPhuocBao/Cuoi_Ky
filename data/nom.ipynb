```python
import pandas as pd
import re
```

```python
df = pd.read_csv('C:/CUOIKY_KPDL/Cuoi_Ky/data/tiktok_videos.csv')
```

```python
def k_to_number(x):
    if pd.isna(x):
        return None

    # Nếu đã là số
    if isinstance(x, (int, float)):
        return float(x)

    if isinstance(x, str):
        x = x.strip().lower()
        x = x.replace(',', '')

        # match số + k / m
        match = re.fullmatch(r'(\d+(\.\d+)?)([km])?', x)
        if match:
            number = float(match.group(1))
            suffix = match.group(3)

            if suffix == 'k':
                return number * 1_000
            elif suffix == 'm':
                return number * 1_000_000
            else:
                return number

    # fallback an toàn
    return None
```

```python
df["like_count"] = df["like_count"].apply(k_to_number)
df["share_count"] = df["share_count"].apply(k_to_number)
df["comment_count"] = df["comment_count"].apply(k_to_number)
```

```python
df.to_csv('tiktok_videos.csv', index=False)
```
